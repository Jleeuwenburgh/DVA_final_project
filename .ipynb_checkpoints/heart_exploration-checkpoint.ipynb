```python
#Checking for and Handling duplicates
n_total = len(df)
n_dup_perfect = df.duplicated().sum()
n_dup_partial = df.duplicated().sum()

print(f"All rows: {n_total} rows")
print(f"Perfect duplicates: {n_dup_perfect}")
print(f"Partial duplicates: {n_dup_partial}")

if n_dup_perfect > 0:
    print("\n Example of duplicate rows:")
    print(df[df.duplicated(keep=False)].sort_values(by=list(df.columns)).head(5))

#Removing perfect duplicates
df = df.drop_duplicates()
n_removed = n_total - len(df)

#Final Check
print(f"\nClead Dataset: {len(df)} righe")
print(f"Perfect duplicates removed: {n_dup_perfect}")
print(f"Partial duplicates remaining: {df.duplicated(subset=['age','sex','cp']).sum()}")
```

The dataset originally contained 1,025 observations.
The analysis of duplicate records shows that there are 723 exact duplicate. 
Before removing the duplicates, partial duplicate analysis was performed to confirm these were true redundancies rather than distinct patients with similar characteristics. All exact duplicates were therefore removed. The cleaned dataset now contains 302 unique patient records.

```python
#Identifying outliers on 'age', 'cp', 'trestbps', 'chol', 'thalach', 'oldpeak','ca' columns
from scipy import stats

cols = ['age', 'cp', 'trestbps', 'chol', 'thalach', 'oldpeak','ca']
df1 = df[cols]
z_scores = stats.zscore(df1)

outliers = np.where(np.abs(z_scores) > 3)[0]
outliers = np.unique(outliers)
df_outliers = df.iloc[outliers]
print("Outliers:\n", df_outliers.head())
print("\nOutliers identified: ", len(df_outliers))
print("Observations: ", len(df))
```

Note: The percentage of outliers in the dataset is approximately 0.04% (13/302). 
As this percentage is practically 0 and below the 5% threshold, they were retained for the subsequent dimensionality reduction.

```python
from sklearn.preprocessing import StandardScaler
#Calculating variance
print("Variance values:\n", df.var(numeric_only=True))

#Standardizing
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[['age', 'cp', 'trestbps', 'chol', 'thalach', 'oldpeak','ca']])
df_scaled_table = pd.DataFrame(df_scaled)
print("\nStandardized columns:\n", df_scaled_table.head())
```

```python
# Performing PCA
from sklearn.decomposition import PCA

#Projecting the data into two principal components
pca = PCA(n_components=2)
components = pca.fit_transform(df_scaled)
pca_df = pd.DataFrame(data=components, columns=['PC1', 'PC2'])

plt.figure(figsize=(12,8))
sns.scatterplot(x='PC1', y='PC2', data=pca_df, hue=df['target'])
plt.title('PCA')
plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.show()

print(pca.explained_variance_ratio_)
print(pca.explained_variance_ratio_.sum())
```

Note: The PCA projection shows a global data structure characterized by strong class overlap with both classes largely mixed across the projection space. The projection does not show clearly defined clusters between patients with and without heart disease, indicating that the linear separability of the dataset is limited.

Moreover, the first two principal components explain 47% of the total variance, which is insufficient to provide an adequate low-dimensional representation.

```python
#PCA on all selected components
pca = PCA(n_components=7)
components = pca.fit_transform(df_scaled)

pca_df = pd.DataFrame(data=components, columns=[f'PC{i+1}' for i in range(7)])
exp_var = pca.explained_variance_ratio_
cum_exp_var = np.cumsum(exp_var)

pca_table = pd.DataFrame({ 'Explained Variance': exp_var,
                         'Cumulative Variance': cum_exp_var}, index=[f'PC{i+1}' for i in range(len(exp_var))])
print(pca_table)
```

Note: Five principal components are needed to explain approximately 80% of the total variance.
Therefore, further analyses will be conducted on these five principal components.

```python
#Applying t-SNE
from mpl_toolkits.mplot3d import Axes3D
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, random_state=0)
data_tsne = tsne.fit_transform(df_scaled)

plt.figure(figsize=(12,8))
sns.scatterplot(x=data_tsne[:,0], y=data_tsne[:,1], hue=df['target'], palette=sns.color_palette("hsv", 2))
plt.title('t-SNE Result')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.show()
```

Note: The t-SNE projection shows strong overlap between the two classes with very few local clusters. However, the global separation between the classes slightly improves compared to PCA, with target 0 concentrated in the lower region of the plot while target 1 shows higher density in the upper region. 
A weak vertical separation pattern emerges, but there is still overlap, which indicates no clearly defined non-linear structure for effective separation.
