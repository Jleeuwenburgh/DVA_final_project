

- **Age**:    age
- **Sex:**    sex
- **cp:**    chest pain type (4 values)
- **trestbps:**    resting blood pressure
- **chol:**    serum cholestoral in mg/dl
- **fbs:**    fasting blood sugar > 120 mg/dl
- **restecg:**    resting electrocardiographic results (values 0,1,2)
- **thalach:**    maximum heart rate achieved
- **exang:**    exercise induced angina
- **oldpeak:**    oldpeak = ST depression induced by exercise relative to rest
- **slope:**    the slope of the peak exercise ST segment
- **ca:**    number of major vessels (0-3) colored by flourosopy
- **thal:**    thal: 0 = normal; 1 = fixed defect; 2 = reversable defect



```python
#Identifying outliers on 'age', 'cp', 'trestbps', 'chol', 'thalach', 'oldpeak','ca' columns
from scipy import stats

cols = ['age', 'cp', 'trestbps', 'chol', 'thalach', 'oldpeak','ca']
df1 = df[cols]
z_scores = stats.zscore(df1)

outliers = np.where(np.abs(z_scores) > 3)[0]
outliers = np.unique(outliers)
df_outliers = df.iloc[outliers]
print("Outliers:\n", df_outliers.head())
print("\nOutliers identified: ", len(df_outliers))
print("Observations: ", len(df))
```

Note: The percentage of outliers in the dataset is approximately 4.8% (49/1025). 
As this percentage is below the 5% threshold, they were retained for the subsequent dimensionality reduction.

```python
from sklearn.preprocessing import StandardScaler
#Calculating variance
print("Variance values:\n", df.var(numeric_only=True))

#Standardizing 
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[['age', 'cp', 'trestbps', 'chol', 'thalach', 'oldpeak','ca']])
df_scaled_table = pd.DataFrame(df_scaled)
print("\nStandardized columns:\n", df_scaled_table.head())
```

```python
# make barplot of sex counts
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='sex')
plt.title('Count of Sex')
plt.show()
```

Note: It seems the dataset is skewed towards one gender

```python
# make a distribution plot of age grouped by sex
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='age', hue='sex', kde=True, bins=30)
plt.title('Age Distribution by Sex')
plt.show()
```

```python
# make correlation heatmap
plt.figure(figsize=(12,10))
corr = df.corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

Note: Not a lot of initial very strong positive or negative correlations found, but there are some

```python
sns.pairplot(df, vars=['age', 'trestbps', 'chol', 'thalach'], hue='target')
plt.suptitle('Pairplot of Age, Trestbps, Chol, Thalach', y=1.02)
plt.show()  
sns.set(style='whitegrid')
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='target', y='chol')
plt.title('Cholesterol Levels by Heart Disease Target')
plt.show()
```

```python
# Performing PCA
from sklearn.decomposition import PCA

#Projecting the data into two principal components
pca = PCA(n_components=2)
components = pca.fit_transform(df_scaled)
pca_df = pd.DataFrame(data=components, columns=['PC1', 'PC2'])

plt.figure(figsize=(12,8))
sns.scatterplot(x='PC1', y='PC2', data=pca_df)
plt.title('PCA')
plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.show()

print(pca.explained_variance_ratio_)
print(pca.explained_variance_ratio_.sum())
```

Note: Two components are not enough because the explained variance is 46%

```python
#PCA on all selected components
pca = PCA(n_components=7)
components = pca.fit_transform(df_scaled)

pca_df = pd.DataFrame(data=components, columns=[f'PC{i+1}' for i in range(7)])
exp_var = pca.explained_variance_ratio_
cum_exp_var = np.cumsum(exp_var)

pca_table = pd.DataFrame({ 'Explained Variance': exp_var,
                         'Cumulative Variance': cum_exp_var}, index=[f'PC{i+1}' for i in range(len(exp_var))])
print(pca_table)
```

Note: Five principal components are needed to explain approximately 80% of the total variance.
Therefore, further analyses will be conducted on these five principal components.
